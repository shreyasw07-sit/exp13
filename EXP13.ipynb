```python
import pandas as pd
import numpy as np

#np.nan for missing values........NaN means not a number
#pd.DataFrame converts it into tabular form

df1=pd.DataFrame({'C1':[11,21, np.nan],'C2':[15, np.nan, np.nan],'C3':[10,20,30]})
df1
```

```python
#To find where is nan
df1.isna()          #Boolean function isna
```

```python
df1.notna()         #this shows NaN as false
```

```python
df1.isna().sum()                                 #tells the number of null values/ NaN in columns
```

```python
df1.isna().sum(axis=1)                           #same thing if we want of row.......axis=1 is rows axis=0 is columns
```

```python
#DATA CLEANING BY DELETING ROWS

df1.dropna()     #NaN Rows are dropped....here axis=0 rows and axis=1 columns
```

```python
#DATA CLEANING BY DELETING COLUMNS

df1.dropna(axis=1)     #NaN Columns are dropped
```

```python
df1.fillna(value='DEFAULT')        #changes NaN to DEFAULT
```

```python
df1.fillna(0)                 # replaced with zero
```

```python
df1.fillna(df1.mean())                #replaced by (mean) column wise
```

```python
#Preprocesing of datasheet 1

import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
df
```

```python
#Replacing '-' with NaN


df.replace("-", np.nan, inplace= True)                #inplace= True replaces the old table with new one
df
```

```python
# as age will be more likely same and student will not have much variation in age so we replace it by "MEAN"
# marks has more variation so replace by "MEDIAN"
# "MODE" is used like if categorical data is there such as department

df.replace("-", np.nan, inplace= True)

#CONVERT COLUMNS TO NUMERIC WHERE NEEDED

df["Age"]= pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]= pd.to_numeric(df["Marks"],errors='coerce')
df
```

```python
# as age will be more likely same and student will not have much variation in age so we replace it by "MEAN"
# marks has more variation so replace by "MEDIAN"
# "MODE" is used like if categorical data is there such as department

#df.replace("-", np.nan, inplace= True)

#CONVERT COLUMNS TO NUMERIC WHERE NEEDED

df["Age"]= pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]= pd.to_numeric(df["Marks"],errors='coerce')
df
```

```python
#Fill missing age with mean and marks with median

df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].median(), inplace=True)
df
```

```python
#changing small department to upper or capital letter

df["Department"]=df["Department"].str.upper()
df
```

```python
# changing date to a same format
#df["Admission_Date"]= pd.to_datetime(df["Admission_Date"], errors='coerce') also true but does not work on colab
df["Admission_Date"]= pd.to_datetime(df["Admission_Date"], format="mixed", dayfirst=True)
(df)
```
